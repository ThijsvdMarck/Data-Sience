```python
import requests
import pandas as pd
```

```python
###Inladen API - kijk naar country code en maxresults
response = requests.get("https://api.openchargemap.io/v3/poi/?output=json&countrycode=NL&maxresults=100&compact=true&verbose=false&key=93b912b5-9d70-4b1f-960b-fb80a4c9c017")
```

```python
###Omzetten naar dictionary
responsejson  = response.json()
# responsejson
```

```python
response.json()
```

```python
# Dataframe bevat kolom die een list zijn. 
# Met json_normalize zet je de eerste kolom om naar losse kolommen
Laadpalen = pd.json_normalize(responsejson)

# Daarna nog handmatig kijken welke kolommen over zijn in dit geval Connections
# Kijken naar eerst laadpaal op de locatie
# Kan je uitpakken middels:
df4 = pd.json_normalize(Laadpalen.Connections)
df5 = pd.json_normalize(df4[0])

# Bestanden samenvoegen
Laadpalen = pd.concat([Laadpalen, df5], axis=1)

# Display the first few rows of the combined dataframe
# Laadpalen.head()
# Save the Laadpalen dataframe as a CSV file
# lp1.to_csv('laadpalen_data.csv', index=False)
```

```python
# Reading the CSV file
laadpalen1 = pd.read_csv('laadpalen_data.csv')

# rename DataFrame to lp1
lp1 = laadpalen1

# Display the first few rows of the dataframe
lp1.head()
```

```python
# Zoek alle ontbrekende waarden in de dataframe
missing_values = lp1.isna().sum()
print(missing_values)
```

```python
# Zoek alle ontbrekende waarden in de dataframe in percentage
missing_values = lp1.isna().sum()
lp1_lenght = len(lp1)
missing_values_percentage = missing_values / lp1_lenght * 100
print(missing_values_percentage)
```

```python
lp1.dropna(subset=['Connections', 'AddressInfo.RelatedURL', 'AddressInfo.ContactTelephone1', 
                   'AddressInfo.ContactTelephone2', 'AddressInfo.AccessComments', 
                   'AddressInfo.ContactEmail', 'AddressInfo.AddressLine2', 'GeneralComments',
                   'StatusTypeID'], 
           inplace=True)
```

```python
# Save the Laadpalen dataframe as a CSV file
lp1.to_csv('laadpalen_datav1.csv', index=False)
```
